```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df_test = pd.read_csv('test.csv')
print(df_test.shape)
df_test.head()
```

```python
df_train = pd.read_csv('train.csv')
print(df_train.shape)
df_train.head()
```

```python
df = pd.read_csv('dataset.csv')
print(df.shape)
df.head()
```

```python
continuous_variables = list(df_train.describe().keys())
continuous_variables.remove('OverallQual')
continuous_variables.remove('OverallCond')
continuous_variables.remove('Id')
##removing the response from predictors list
continuous_variables.remove('SalePrice')
```

```python
categorical_variables = list(set(df_train.columns)-set(continuous_variables))
categorical_variables.remove('Id')
categorical_variables.remove('SalePrice')
```

```python
print('categorical_variables\t', len(categorical_variables), '\continuous_variables\t', len(continuous_variables))
```

```python
missing_percent = 0
variables_to_delete = []
```

```python
categorical_missing_variables = []
for cat in categorical_variables:
    if df[cat].isnull().sum() > 0:
        missing_percent = round(df[cat].isnull().sum()/df.shape[0], 4)*100
        categorical_missing_variables.append(cat)
        print(cat+'\t\t', str(df[cat].isnull().sum())+'\t', missing_percent)
        if missing_percent > 45:
            variables_to_delete.append(cat)
```

```python
continuous_missing_variables = []
for cat in continuous_variables:
    if df[cat].isnull().sum() > 0:
        missing_percent = round(df[cat].isnull().sum()/df.shape[0], 4)*100
        continuous_missing_variables.append(cat)
        print(cat+'\t\t', str(df[cat].isnull().sum())+'\t', missing_percent)
        if missing_percent > 45:
            variables_to_delete.append(cat)
```

```python
print(variables_to_delete)
df.drop(labels = variables_to_delete, axis = 1, inplace = True)
df.drop(labels = ['Id'], axis = 1, inplace = True)
categorical_variables = list(set(categorical_variables) - set(variables_to_delete))
categorical_missing_variables = list(set(categorical_missing_variables) - set(variables_to_delete))
continuous_variables = list(set(continuous_variables) - set(variables_to_delete))
continuous_missing_variables = list(set(continuous_missing_variables) - set(variables_to_delete))
```

```python
from statistics import mode
```

```python
imputed_value = 0
for cat in continuous_missing_variables.copy():
    missing_percent = round(df[cat].isnull().sum()/df.shape[0], 4)*100
    if missing_percent <= 2:
        continuous_missing_variables.remove(cat)
        if np.mean(df[cat]) > 10:
            imputed_value = np.mean(df[cat])
        else:
            imputed_value = mode(df[cat])
        df[cat].fillna(imputed_value, inplace = True)
        #print(cat+'\t', str(imputed_value)+'\t', missing_percent)
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
```

```python
for cat in continuous_missing_variables.copy():
    y = cat
    x = list(set(continuous_variables) - set(continuous_missing_variables))
    continuous_missing_variables.remove(cat)
    train = df[df[y].notnull()]
    test = df[df[y].isnull()]
    x_train, x_test, y_train, y_test = train[x], test[x], train[y], test[y]
    linear_reg.fit(x_train, y_train)
    df.loc[df[y].isnull(),[y]] = linear_reg.predict(x_test)
    #print(y)
    #print(x)
    #print('\n')
```

```python
categorical_non_missing_variables = list(set(categorical_variables)-set(categorical_missing_variables))
```

```python
import category_encoders as en
bin_encoding = en.BinaryEncoder()
df_category = bin_encoding.fit_transform(df[categorical_non_missing_variables])
df_full = pd.concat([df[continuous_variables], df_category], axis = 1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 50, n_jobs = -1)
```

```python
for cat in categorical_missing_variables.copy():
    y = cat
    x = df_full.columns
    categorical_missing_variables.remove(cat)
    temp = pd.concat([df_full, df[y]], axis = 1)
    train = temp[temp[y].notnull()]
    test = temp[temp[y].isnull()]
    x_train, x_test, y_train, y_test = train[x], test[x], train[y], test[y]
    knn.fit(x_train, y_train)
    df.loc[df[y].isnull(),[y]] = knn.predict(x_test)
    bin_encoding = en.BinaryEncoder()
    df_category = bin_encoding.fit_transform(df[y])
    df_full = pd.concat([df_full, df_category], axis = 1)
    #print(y)
    #print(x)
    #print('\n')
```

```python
predict_response_cleaned = pd.concat([df_full, df_train.SalePrice], axis = 1)
```
